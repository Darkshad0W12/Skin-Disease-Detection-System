```python
from keras.models import load_model
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import numpy as np
```

```python
model = load_model("C:/Users/ayush/OneDrive/Desktop/mobilenet_xception.keras")
```

```python
img = load_img("C:/Users/ayush/OneDrive/Desktop/projectpics/train/Psor/DIMG-5521.jpg",target_size=(224,224))
#img = load_img("C:/Users/ayush/OneDrive/Desktop/projectpics/val/NonPsor/erythema-multiforme-34.jpg",target_size=(224,224))
img = img_to_array(img)
plt.imshow(img/225)
plt.show()
img = np.expand_dims(img,0)
result = float(model.predict(img)[0])
if result<0.7:
    print("Psoriasis", result)
else:
    print("Non Psoriasis", result)
```
